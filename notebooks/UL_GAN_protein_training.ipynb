```python
from model_graph import RGDiscrimininator, RGULGenerator, GANTrainerComb
```

```python
device = 'cuda:4'
generator = RGULGenerator(in_dim=128, initial_dim=128, \
            hidden_dims=[[64, 128], [128, 128], [128, 256]], \
            final_layer_dims=[64], hidden_act='relu', \
            edge_hidden_dim=16, \
            leaky_relu_coef=0.05, device=device, \
            skip_z=True, skip_z_dims=None, \
            unpool_bn=False, link_bn=False, \
            link_act='leaky', unpool_para=dict(add_additional_link=True, add_perference=True, \
                                roll_bn=False, roll_simple=True, add_bn=False), \
            attr_bn=False, edge_out_dim=None, \
            fix_points=[2, 0], roll_ones=[0, 0], node_nums = [3, 4, 8], \
            use_bn=False, final_bn=False, node_feature=3
            , skipz_bn=False
            )
discriminator = RGDiscrimininator(in_dim=3, hidden_dim=[128,256], lin_hidden_dim=128, 
                out_hidden_dim=[256, 256], device=device, \
                useBN=False, droprate=None, outBN=False, out_drop=0.3, \
                final_layers=2, \
                conv_layers=2,
                last_act='sigmoid', relu_coef=0.05)

generator = generator.to(device)
discriminator = discriminator.to(device)
```

```python
import torch
real_data = torch.load('protein_train.pt')
```

```python
from torch_geometric.loader import DataLoader
batch_size = 64
data_loader = DataLoader(real_data, batch_size=batch_size, shuffle=True, follow_batch=['edge_index'])
```

```python
for j in data_loader:
    break
```

```python
trainer = GANTrainerComb(discriminator, generator, rand_dim=128, train_folder='protein_data', \
                   tot_epoch_num=500, eval_iter_num=100, batch_size=64, \
                   device=device, d_add=None, learning_rate_g=2e-5, learning_rate_d=1e-4, \
                   lambda_g=0, max_train_G=20, tresh_add_trainG=0.2, \
                   use_loss='bce', \
                   g_out_prob=True, lambda_rl=2e-1, \
                   lambda_nonodes = 0.,
                   lambda_noedges = 0.,
                   trainD=True, \
                   initial_weight=True
            )
```

```python
trainer.train(data_loader, verbose=False, \
        NN=200, evaluate_num=1000, mol_data=j, \
        alter_trainer=False, reinforce_acclerate=True, only_train=None)
```
